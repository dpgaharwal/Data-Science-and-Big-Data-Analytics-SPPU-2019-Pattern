```python
#Calculate the mean of the stroke column
stroke_mean = df["stroke"].astype("float").mean(axis = 0)
print("The Mean value of stroke column: " ,stroke_mean)

# replace NaN by mean value in "stroke" column
df["stroke"].replace(np.nan,stroke_mean,inplace = True)
```

```python
# Calculate the mean value for the 'horsepower' column
avg_hp = df["horsepower"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_hp)

# replace NaN by mean value in "stroke" column
df["horsepower"].replace(np.nan, avg_hp, inplace = True)
```
