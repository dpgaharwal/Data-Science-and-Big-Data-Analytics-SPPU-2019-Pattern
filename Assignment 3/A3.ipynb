```python
# Group data by species
grouped = iris.groupby('Species')

# List of species to display statistics for
species_list = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

# Loop through each species and display statistics
for species in species_list:
    print(f"Statistics for {species}:")
    species_data = grouped.get_group(species)
    print(species_data.describe())
    print("\n")
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris,hue = 'Species',height = 7).map(plt.scatter, "SepalLengthCm","SepalWidthCm").add_legend()
```
